## Celestial Coordinates

### RA and Dec

Astronomers measure positions on the sky using equatorial coordinates: right ascension ($\alpha$) and declination ($\delta$). You can think of as longitude and latitude on the celestial sphere, relative to the celestial equator and the celestial north and south pole.

Right ascension and declination (like latitude and longitude) are both angles.  Declination is measured in degrees, minutes, and seconds from the celestial equator, just like latitude.  Right ascension is measured in hours, minutes, and seconds, where 24 hours is the full 360Â° circle.  

The first thing you will want to do is create a function to convert RA and Dec into decimal degrees.


```python
def ra(ra_hours, ra_minutes, ra_seconds):
    # Convert right ascension to decimal degrees
    ra_decimal = (ra_hours + ((ra_minutes + ra_seconds) / 60)) * 360 / 24
    return ra_decimal

def dec(dec_degrees, dec_minutes, dec_seconds):
    # Convert declination to decimal degrees
    dec_decimal = dec_degrees + (dec_minutes + dec_seconds / 60) / 60
    return dec_decimal

# isnt it supposed to be 3600?
```
