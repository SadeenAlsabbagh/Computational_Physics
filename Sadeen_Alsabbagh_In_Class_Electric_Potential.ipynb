# Introduction to the Laplace Equation



The gradient is a vector operator, defined as:
$\nabla=\frac{\partial}{\partial x} \hat{x}+ \frac{\partial}{\partial y} \hat{y}+ \frac{\partial}{\partial z} \hat{z}.$

Another way to find the gravitational force on an object is to find $\vec{F}=-\nabla U$.

For $U=mgy$,
\begin{eqnarray}\vec{F}&=&-\nabla U\\
    &=&-\left(\frac{\partial}{\partial x} \hat{x}+ \frac{\partial}{\partial y} \hat{y}+ \frac{\partial}{\partial z} \hat{z}\right)U\\
    &=&-\left(\frac{\partial U}{\partial x} \hat{x}+ \frac{\partial U}{\partial y} \hat{y}+ \frac{\partial U}{\partial z} \hat{z}\right)
\end{eqnarray}

Since $U$ depends only on $y$, all terms are zero except for
\begin{eqnarray}
    \vec{F}&=&-\frac{\partial U}{\partial y} \hat{y}\\
    &=&-\frac{\partial (mgy)}{\partial y} \hat{y}\\
    &=&- mg \hat{y}\\
\end{eqnarray}

Now, thinking in terms of our gravitational field $\vec{\Gamma}$,
\begin{eqnarray}
m\vec{\Gamma} &=& \vec{F}\\
\vec{\Gamma} &=& \frac{\vec{F}}{m}
\\
 &=& \frac{-\nabla{U}}{m} \\
  &=& \frac{-\nabla{(mgy)}}{m} \\
    &=& -\nabla(gy)\\
        &=& -\nabla V
\end{eqnarray}
as we defined the potential $V$ before.

Now, consider electric field, which exerts a force $\vec{F}$ on a particle with a charge $q$:
\begin{eqnarray}
    \vec{E}=\frac{\vec{F}}{q}.
\end{eqnarray}
and we know $\vec{F}=-\nabla U$ because the electric force is conservative, so:
\begin{eqnarray}
    \vec{E}=\frac{-\nabla{U}}{q}.
\end{eqnarray}
We divided the gravitational potential energy by $m$ to get the gravitational potential, because the gravitational force depends on the mass of the object feeling the force.  But if we let the electric potential be $V=\frac{U}{q}$, then
\begin{eqnarray}
    \vec{E}=-\nabla{V}.
\end{eqnarray}
Like $\vec{\Gamma}$, the gravitational field, $\vec{E}$ tells us the direction of the force that will act on the charged objects in the field, and the strength of the force is $F=qE$ with the added complication that, unlike $m$, $q$ could be positive or negative. Negatively charged particles in the $\vec{E}$ field will travel the opposite direction to positive charges.


Since $\vec{E}=-\nabla V=$,
\begin{eqnarray}
\vec{E}&=&-\left(\frac{\partial}{\partial x} \hat{x}+ \frac{\partial}{\partial y} \hat{y}+ \frac{\partial}{\partial z} \hat{z}\right)V\\
&=&-\frac{\partial V}{\partial x} \hat{x}-\frac{\partial V}{\partial y} \hat{y}-\frac{\partial V}{\partial z} \hat{z}
\end{eqnarray}
so $E_x=-\frac{\partial V}{\partial x}$, $E_y=-\frac{\partial V}{\partial y}$, and $E_z=-\frac{\partial V}{\partial z}$.

This means that the force will depend on the ``slope" of V in each direction.  In three dimensions, that's a bit hard to imagine.  It's easier to visualize if $V=V(x)$ or $V=V(x,y)$, and does not vary with $z.$

In Physics 102, you may have or will learn about Gauss's Law, which describes how an electric field is produced by electric charges.  In 102, we learn the integral form:
\begin{eqnarray}
\oint{\vec{E}\cdot d\vec{A}}=\frac{Q_{enc}}{\varepsilon_0}=\frac{1}{\varepsilon_0}\int{\rho}~dV.
\end{eqnarray}
where $\rho$ is the charge density inside a closed surface, and the total charge inside the surface is $Q_{enc}$. The electric field along the surface is $\vec{E}$.
There is also a derivative form that we learn in Physics 340:
\begin{eqnarray}
\nabla\cdot\vec{E}=\frac{\rho}{\epsilon_o},
\end{eqnarray}
where $\nabla\cdot\vec{E}$ is the gradient of the electric field,
\begin{eqnarray}
    \nabla\cdot\vec{E}&=&\left(\frac{\partial}{\partial x} \hat{x}+ \frac{\partial}{\partial y} \hat{y}+ \frac{\partial}{\partial z} \hat{z}\right)\cdot \vec{E}\\
    &=&\left(\frac{\partial}{\partial x} \hat{x}+ \frac{\partial}{\partial y} \hat{y}+ \frac{\partial}{\partial z} \hat{z}\right)\cdot (E_x \hat{x}+ E_y\hat{y}+E_z\hat{z})\\
        &=&\frac{\partial E_x}{\partial x} \hat{x}\cdot\hat{x} + \frac{\partial E_y}{\partial y} \hat{y}\cdot\hat{y}+ \frac{\partial E_z}{\partial z} \hat{z}\cdot\hat{z}\\
         &=&\frac{\partial E_x}{\partial x}  + \frac{\partial E_y}{\partial y} + \frac{\partial E_z}{\partial z}
\end{eqnarray}
We now have the mathematical equipment to deal with something much more complex situation than our simple case of the gravitational field, caused by the spherical Earth below us.

We can find how the electric field varies based on some lumpy distribution of charge given by a charge density $\rho(x,y,z)$.

Recall that $E_x=-\frac{\partial V}{\partial x}$, $E_y=-\frac{\partial V}{\partial y}$, and $E_z=-\frac{\partial V}{\partial z}$.  Plugging this in to Gauss's Law:

\begin{eqnarray}
\frac{\partial E_x}{\partial x}  + \frac{\partial E_y}{\partial y} + \frac{\partial E_z}{\partial z} &=& \frac{\rho} {\varepsilon_o}\\
\frac{\partial}{\partial x}\left(-\frac{\partial V}{\partial x}\right)  + \frac{\partial }{\partial y} \left(-\frac{\partial V}{\partial y}\right) + \frac{\partial }{\partial z}\left(-\frac{\partial V}{\partial z}\right)  &=& \frac{\rho} {\varepsilon_o}\\
\frac{\partial^2 V}{\partial x^2} + \frac{\partial^2 V}{\partial y^2} + \frac{\partial^2 V}{\partial z^2}  &=& -\frac{\rho} {\varepsilon_o}
\end{eqnarray}
This is Poisson's Equation. You can also arrive at this by taking $\nabla\cdot\vec{E}=-\nabla\cdot\nabla V$ and working out $\nabla\cdot\nabla=\nabla^2$, which is called the ''Laplace operator" or "Laplacian", and applying it to $V$.  

While we could imagine many complicated distributions of charges, another interesting situation is where $\rho=0$ in some area, but is not necessarily 0 outside the area. In this situation, we have to solve the second-order partial differential equation:
\begin{eqnarray}
    \frac{\partial^2 V}{\partial x^2} + \frac{\partial^2 V}{\partial y^2} + \frac{\partial^2 V}{\partial z^2}  &=& 0
\end{eqnarray}
This is called Laplace's Equation, or Poisson's Equation in a vacuum.  It may seem boring, but remember that when solving a differential equation, you must match the boundary conditions, and arrangements of charges outside the vacuum area could certainly create an interesting potential.


# The Method of Relaxation


Laplace's equation for V can be solved numerically using something called the Method of Relaxation.

First, consider the simple case that $V=V_o$ at $x=0$, and $V=0$ at $x=a$.  You can imagine creating this situation with a large charged capacitor with its positively-charged plate in the $x=0$ plane and its negatively-charged plate in the $x=a$ plane, connected to ground.

As long as we avoid the edges of the capacitor, $V$ will not depend on $y$ or $z$, only $x$.

A numerical approximation to $\frac{dV}{dx}\approx\frac{\Delta V}{\Delta x}.$
Let's consider the points $x+h$ and $x-h$.  

Then
\begin{eqnarray}
\left.\frac{dV}{dx}\right|_x &\approx& \frac{V(x+h)-V(x-h)}{h} \\
\end{eqnarray}
and the numerical approximation to the second derivative is:
\begin{eqnarray}
\frac{d^2 V}{d x^2}&\approx& \frac{ \Delta\left(\frac{dV}{dx}\right)}
{\Delta x}\\
\left. \frac{d^2 V}{d x^2}\right|_x &\approx& \frac{ \left. \frac{dV}{dx}\right|_{x+h}-\left. \frac{dV}{dx}\right|_{x-h}}{h}\\
&\approx& \frac{\frac{V(x+h+h)-V(x+h-h)}{h}-\frac{V(x-h+h)-V(x-h-h)}{h}}{h} \\
&=& \frac{V(x+2h)-V(x)-[V(x)-V(x-2h)]}{h^2}\\
&=& \frac{V(x+2h)-V(x)-V(x)+V(x-2h)}{h^2}\\
&=& \frac{V(x+2h)-2V(x)+V(x-2h)}{h^2}\\
\end{eqnarray}

Laplace's equation says that $\frac{d^2 V}{dx^2}=0$, sp we can take
\begin{eqnarray}
0&=& \frac{V(x+2h)-2V(x)+V(x-2h)}{h^2}\\
0&=& V(x+2h)-2V(x)+V(x-2h)\\
2V(x)&=& V(x+2h)+V(x-2h)\\
V(x)&=& \frac{1}{2}[V(x+2h)+V(x-2h)]\\
\end{eqnarray}

Since $h$ is just some arbitrary $\Delta x$, let's change that to
\begin{eqnarray}
V(x)= \frac{1}{2}[V(x+h)+V(x-h)]
\end{eqnarray}
which you might recognize just says that $V(x)$ is approximately the average of $V(x+h)$ and $V(x-h)$.

This is the equation we can use to find V(x) numerically.  

# The Numerical Solution of the Laplace Equation

Begin by setting $a=10$ m and $h=1$.  

Then define a grid of x values with a separation of $h$ between each point.  

Then make an array filled with zeros to hold $V(x)$.  

As when we solve any differential equation, we must specify the boundary conditions. Let $V_o=6$ V. Place this into the appropriate place in the array V.

```python
import numpy as np

# Step 1: Define parameters
a = 10  # m
h = 1   # grid separation

# Step 2: Define grid of x values
x_values = np.arange(0, a+h, h)

# Step 3: Initialize array for V(x) filled with zeros
V = np.zeros_like(x_values)

# Step 4: Specify and assign boundary condition V0 = 6 V
V[0] = 6

# Print the initial values for verification
print("x values:", x_values)
print("Initial V(x) values:", V)
```

## Experimenting with few data points.

Now plot our current "approximation" of V(x).

Plot $V(0)$ and $V(a)$ with points so we remember that these are our boundary values.

```python
import matplotlib.pyplot as plt

# Plot V(x)
plt.plot(x_values, V, marker='o', label='Approximation of V(x)')
# Mark boundary values
plt.plot(0, V[0], marker='o', color='red', label='V(0)', markersize=8)
plt.plot(a, V[-1], marker='o', color='blue', label='V(a)', markersize=8)

# Labeling and title
plt.xlabel('x')
plt.ylabel('V(x)')
plt.title('Approximation of V(x) with Boundary Values')
plt.legend()
plt.grid(True)

# Show plot
plt.show()
```

Now we need to apply what we learned above.  

**Without changing the first and last value in the V array**, let's create a new V by averaging each point with the point beside it.  You can do this with a for loop (easy to program and easy for the person who reads your code to understand) or with an array operation (hard to program and read, but fast).

Then plot the result.

```python
# Create a new array for updated V(x)
V_new = np.copy(V)  # Copy the original V array

# Iterate over each point (excluding the first and last)
for i in range(1, len(V)-1):
    # Average each point with the point beside it
    V_new[i] = (V[i-1] + V[i+1]) / 2

# Plot the updated approximation of V(x)
plt.plot(x_values, V_new, marker='o', label='Updated Approximation of V(x)')
# Mark boundary values
plt.plot(0, V_new[0], marker='o', color='red', label='V(0)', markersize=8)
plt.plot(a, V_new[-1], marker='o', color='blue', label='V(a)', markersize=8)

# Labeling and title
plt.xlabel('x')
plt.ylabel('V(x)')
plt.title('Updated Approximation of V(x) with Boundary Values')
plt.legend()
plt.grid(True)

# Show plot
plt.show()
```

# Relaxation Function

```python
# Define a function for one iteration of relaxation
def relaxation_one_iteration(V):
    """
    Perform one iteration of relaxation to solve Laplace equation.

    Parameters:
        V (ndarray): Initial guess of V(x).

    Returns:
        V_new (ndarray): Updated approximation of V(x).
    """
    V_new = np.copy(V)
    for i in range(1, len(V) - 1):
        V_new[i] = 0.5 * (V[i-1] + V[i+1])

    return V_new

# Perform one iteration of relaxation
V_relaxed_one_iteration = relaxation_one_iteration(V)

# Plot the result
plt.plot(x_values, V_relaxed_one_iteration, marker='o', label='One Iteration of Relaxation')
plt.plot(0, V_relaxed_one_iteration[0], marker='o', color='red', label='V(0)', markersize=8)
plt.plot(a, V_relaxed_one_iteration[-1], marker='o', color='blue', label='V(a)', markersize=8)

# Labeling and title
plt.xlabel('x')
plt.ylabel('V(x)')
plt.title('One Iteration of Relaxation Method')
plt.legend()
plt.grid(True)

# Show plot
plt.show()
```

```python
# Define a function for relaxation iteration
def relaxation_iteration(V, tol=1e-6, max_iter=1000):
    """
    Perform relaxation iteration to solve Laplace equation.

    Parameters:
        V (ndarray): Initial guess of V(x).
        tol (float): Tolerance for convergence.
        max_iter (int): Maximum number of iterations.

    Returns:
        V_new (ndarray): Updated approximation of V(x).
        num_iter (int): Number of iterations performed.
    """
    V_new = np.copy(V)
    num_iter = 0
    diff = tol + 1  # Initialize difference to ensure at least one iteration

    while diff > tol and num_iter < max_iter:
        V_old = np.copy(V_new)
        for i in range(1, len(V) - 1):
            V_new[i] = 0.5 * (V_old[i-1] + V_old[i+1])
        diff = np.max(np.abs(V_new - V_old))
        num_iter += 1

    return V_new, num_iter

# Perform relaxation iteration
V_relaxed, num_iterations = relaxation_iteration(V)

# Plot the result
plt.plot(x_values, V_relaxed, marker='o', label='Relaxed Approximation of V(x)')
plt.plot(0, V_relaxed[0], marker='o', color='red', label='V(0)', markersize=8)
plt.plot(a, V_relaxed[-1], marker='o', color='blue', label='V(a)', markersize=8)

# Labeling and title
plt.xlabel('x')
plt.ylabel('V(x)')
plt.title('Relaxed Approximation of V(x) with Boundary Values')
plt.legend()
plt.grid(True)

# Show plot
plt.show()

print("Number of iterations performed:", num_iterations)


```

Hmmm.  One iteration doesn't do much.  Let's make that thing up there into a function, and then run it 5 times.

```python
# Define a function for multiple iterations of relaxation
def relaxation_multiple_iterations(V, num_iterations=5):
    """
    Perform multiple iterations of relaxation to solve Laplace equation.

    Parameters:
        V (ndarray): Initial guess of V(x).
        num_iterations (int): Number of iterations to perform.

    Returns:
        V_new (ndarray): Updated approximation of V(x) after multiple iterations.
    """
    V_new = np.copy(V)
    for _ in range(num_iterations):
        V_new = relaxation_one_iteration(V_new)
    return V_new

# Perform 5 iterations of relaxation
V_relaxed_5_iterations = relaxation_multiple_iterations(V, num_iterations=5)

# Plot the result
plt.plot(x_values, V_relaxed_5_iterations, marker='o', label='5 Iterations of Relaxation')
plt.plot(0, V_relaxed_5_iterations[0], marker='o', color='red', label='V(0)', markersize=8)
plt.plot(a, V_relaxed_5_iterations[-1], marker='o', color='blue', label='V(a)', markersize=8)

# Labeling and title
plt.xlabel('x')
plt.ylabel('V(x)')
plt.title('5 Iterations of Relaxation Method')
plt.legend()
plt.grid(True)

# Show plot
plt.show()
```

## A Reasonable Solution

Hmm.  The extra 100 doesn't seem to help much.  How can we tell when the algorithm is done? Let's go back to the function and see if we can come up with a way to quantify how good the approximation is.

Then we could use a while loop to run the function over and over until it has returned a good appoximation.

```python
def relaxation_until_convergence(V, tol=1e-6, max_iter=1000):
    """
    Perform relaxation iteration until convergence to solve Laplace equation.

    Parameters:
        V (ndarray): Initial guess of V(x).
        tol (float): Tolerance for convergence.
        max_iter (int): Maximum number of iterations.

    Returns:
        V_final (ndarray): Approximation of V(x) after convergence.
        num_iter (int): Number of iterations performed.
    """
    V_new = np.copy(V)
    num_iter = 0
    diff = tol + 1  # Initialize difference to ensure at least one iteration

    while diff > tol and num_iter < max_iter:
        V_old = np.copy(V_new)
        for i in range(1, len(V) - 1):
            V_new[i] = 0.5 * (V_old[i-1] + V_old[i+1])
        diff = np.max(np.abs(V_new - V_old))
        num_iter += 1

    return V_new, num_iter

# Initial guess for V(x)
V_initial_guess = np.zeros_like(x_values)
V_initial_guess[0] = 6  # Boundary condition V(0) = 6 V

# Perform relaxation iteration until convergence
V_final, num_iterations = relaxation_until_convergence(V_initial_guess)

# Plot the result
plt.plot(x_values, V_final, marker='o', label='Relaxed Approximation of V(x)')
plt.plot(0, V_final[0], marker='o', color='red', label='V(0)', markersize=8)
plt.plot(a, V_final[-1], marker='o', color='blue', label='V(a)', markersize=8)

# Labeling and title
plt.xlabel('x')
plt.ylabel('V(x)')
plt.title('Relaxed Approximation of V(x) with Boundary Values')
plt.legend()
plt.grid(True)

# Show plot
plt.show()

print("Number of iterations performed:", num_iterations)
```

There we go.  We didn't need hundreds of loops to get the job done.

The program looks fine, but we should check the result and make sure the physics is right.  We know that $\vec{E}=-\nabla V=-\frac{dV}{dx}\hat{x}$  in this case.

We could code up the numerical approximation to the first derivative, but let's just use numpy, eh?


```python
import numpy as np
import matplotlib.pyplot as plt

# Calculate electric field using numpy gradient
Ex = -np.gradient(V_final, x_values)

# Plot the electric field
plt.plot(x_values, Ex)
plt.title('Electric Field in a Capacitor')
plt.xlabel('x (m)')
plt.ylabel('E (N/C)')
plt.grid(True)
plt.show()
```

It's a bit wobbly at the ends, but that's to be expected taking the numerical derivative with so few points.  Let's go back and recalculate everything with more points and see if that's nicer.

```python
import numpy as np
import matplotlib.pyplot as plt

# Define parameters
a = 10  # m
h = 0.1   # grid separation

# Define grid of x values with more points
x_values = np.arange(0, a+h, h)

# Initial guess for V(x)
V_initial_guess = np.zeros_like(x_values)
V_initial_guess[0] = 6  # Boundary condition V(0) = 6 V

# Perform relaxation iteration until convergence
V_final, num_iterations = relaxation_until_convergence(V_initial_guess)

# Calculate electric field using numpy gradient with more points
Ex = -np.gradient(V_final, x_values)

# Plot the result
plt.plot(x_values, Ex)
plt.title('Electric Field in a Capacitor')
plt.xlabel('x (m)')
plt.ylabel('E (N/C)')
plt.grid(True)
plt.show()

print("Number of iterations performed:", num_iterations)
```
